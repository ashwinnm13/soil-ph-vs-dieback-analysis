```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
```

```python
# Soil pH (x) and Percentage Dieback (y)
soil_ph = np.array([3.3, 3.4, 3.4, 3.5, 3.6, 3.6, 3.7, 3.7, 3.8, 3.8])
dieback = np.array([7.3, 10.8, 13.1, 10.4, 5.8, 9.3, 12.4, 14.9, 11.2, 8.0])
```

```python
plt.figure()
plt.scatter(soil_ph, dieback)
plt.xlabel("Soil pH")
plt.ylabel("Percentage Dieback")
plt.title("Scatter Plot: Soil pH vs Crown Dieback")
plt.show()
```

```python
#Correlation Analysis
corr, _ = pearsonr(soil_ph, dieback)
print("Pearson Correlation Coefficient:", corr)
```

```python
# Reshape for sklearn
X = soil_ph.reshape(-1, 1)
y = dieback

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
# Predicted values
y_pred = model.predict(X)

plt.figure()
plt.scatter(soil_ph, dieback)
plt.plot(soil_ph, y_pred)
plt.xlabel("Soil pH")
plt.ylabel("Percentage Dieback")
plt.title("Linear Regression: Soil pH vs Dieback")
plt.show()
```

```python
r2 = model.score(X, y)
print("R² value:", r2)
# The relationship between soil pH and dieback is very weak
# Soil pH alone is not a strong predictor of crown dieback in this dataset
```
